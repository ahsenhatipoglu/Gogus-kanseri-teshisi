<h2>1. Giriş</h2>

Meme kanseri, kadınlar arasında ikinci ölüm nedenidir. Erken teşhis ve ardından uygun kanser tedavisi ölümcül riski azaltabilir. Tümörler iyi veya kötü huylu olabilir. İyi huylu tümörler kanser değildir. Meme kanseri kötü huylu bir tümördür ve göğüs hücrelerinde gelişir.

Tıp uzmanları bir hastalığı tespit ederken hatalar yapabilirler. Veri madenciliği ve makine öğrenimi gibi teknolojinin yardımı, teşhis doğruluğunu önemli ölçüde artırabilir. Yapay Sinir Ağları (YSA), akıllı meme kanseri teşhisinde yaygın olarak kullanılmaktadır.


<h2>2.(Artificial Neural Network (ANN))</h2>

Yapay sinir ağları insan beyni örnek alınarak, öğrenme sürecinin matematiksel olarak modellenmesi sonucu ortaya çıkmıştır. Beyindeki biyolojik sinir ağlarının yapısını, öğrenme, hatırlama ve genelleme kabiliyetlerini taklit eder.

<b>drop()</b> fonksiyonu istenilen satırların silindiği yeni bir DataFrame döndürür. Yani asıl DataFrame(df)’de bir değişiklik olmaz. Bu problemi çözmek için 2 alternatif vardır:

Aşağıdaki kod satırında, meme kanserini tahmin etmek için gerekli olmayan <b>"id"</b> ve <b>"Unnamed: 32"</b> sütunlarının silindiği görülmektedir.

```python
# Gereksiz sütunların silinmesi ('id' ve 'Unnamed: 32')
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
# Verinin ilk 5 satırını görüntüleme
df.head(5)
```

```python
# 'diagnosis' sütununu kategorik ifadelerden (M/B) sayısal verilere (1/0) dönüştürme
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
y=df['diagnosis']
X=df.drop('diagnosis',axis=1)
```

```python
# Özelliklerin dağılımını histogramlarla görselleştirme
df.hist(figsize = (30,30), color = 'orange')
plt.show()
```

```python
# Korelasyon matrisinin hesaplanması ve görselleştirilmesi
df_numeric = df.select_dtypes(include=[np.number])  # Sayısal sütunları seç
korelasyon = df_numeric.corr() 

plt.figure(figsize=(20,20))
sns.heatmap(korelasyon, cbar=True, square= True, fmt='.1f', annot=True, annot_kws={'size':15}, cmap='coolwarm')
plt.show()
```

<p></p>
y'deki <b>M</b> ve <b>B</b> değerleri <b>1</b> ve <b>0</b> olarak değiştirilmektedir. Buradaki M değeri kötü huylu tümörü, B değeri ise iyi huylu tümörü temsil etmektedir.

```python
y=y.replace(['M'],1)
y=y.replace(['B'],0)
```

```python
y
```

<p></p>
Verilen veriler <b>sklearn</b> kullanılarak eğitim verileri ve test verileri olarak bölünmektedir.

```python
# Eğitim ve test verisinin oluşturulması
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33)
```

<p></p>
Uygulamada veri kümesi <b>sklearn</b> kullanılarak eğitim aşaması için <b>%66.7</b> ve test aşaması için <b>%33.3</b> olacak şekilde bölünmektedir.

<p></p>
<b>Özellik standardizasyonu</b> veya <b>Z-skor normalizasyonu</b> yöntemi ile veriler ölçeklendirilir. Bu işlem için Python'da <b>sklearn.preprocessing</b> kütüphanesinin  <b>StandardScaler</b> metodu  içe aktarılıp kullanılmaktadır. Standardizasyon, ortalama değerin 0, standart sapmanın ise 1 değerini aldığı, dağılımın normale yaklaştığı bir metottur.
Eldeki değerden ortalama değeri çıkartarak, sonrasında varyans değerine bölerek hesaplanmaktadır.

<b>StandardScaler</b> ortalamayı kaldırır ve her bir özelliği/değişkeni birim varyansına ölçeklendirir. Bu işlem, özellik açısından bağımsız bir şekilde gerçekleştirilir.

Özelliklerin standart sapmasının 1 ile 0 civarında ortalanacak şekilde standartlaştırılması, farklı birimleri olan ölçümler karşılaştırıldığında değil, aynı zamanda makine öğrenme algoritması için de genel bir gereklilik olarak önemlidir. Dereceli azalma önemli bir örnek olarak düşünülebilir. StandardScaler lojistik regresyonda, destek vektör makinelerinde (SVM) ve sinir ağlarında sıklıkla kullanılan bir optimizasyon algoritmasıdır.


```python
# Özelliklerin standartlaştırılması
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
N, D = X_train.shape
```

<p></p>
Bu sınıflandırma problemi için bir <b>Keras 2.0</b> modeli oluşturulmaktadır.

```python
# Yapay sinir ağı modeli oluşturma
import tensorflow as tf 
from tensorflow import keras

model = tf.keras.models.Sequential([
  tf.keras.layers.Input(shape=(D,)),
  tf.keras.layers.Dense(1, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
    
])
```

```python
# Modelin derlenmesi
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Modelin eğitilmesi
r = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=150)
```

Tüm derin öğrenme modellerini eğitirken kaydedilen varsayılan geri aramalardan biri <b>history()</b> aramadır. Her dönem için eğitim ölçümlerini kaydeder. Bu, eğer ayarlanmışsa, doğrulama veri kümesinin kaybı ve doğruluğunun yanı sıra(sınıflandırma sorunları için) kaybı ve doğruluğunu içerir.

<b>history</b> nesnesi, modeli eğitmek için kullanılan fit() işlevine yapılan çağrılardan döndürülür. Metrikler, döndürülen nesnenin geçmiş üyesindeki bir sözlükte saklanır.


```python
# Eğitim ve doğrulama kayıplarının görselleştirilmesi
import matplotlib.pyplot as plt
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
```

```python
# Model kayıplarının DataFrame'e dönüştürülmesi 
model_loss = pd.DataFrame(model.history.history)
model_loss
```

```python
# Model kayıplarının DataFrame'e görselleştirilmesi
model_loss.plot()
```

<p></p>
<b>predictions</b> üzerinde tahminler oluşmaktadır. 

```python
# Modelin tahminlerini alma ve ikili sınıflandırma için sonucu 0 veya 1'e dönüştürme
predictions = model.predict(X_test)

# Sınıf tahminlerini 0 veya 1'e dönüştür (ikili sınıflandırma için)
predictions = (predictions > 0.5).astype("int32")
```

```python
# Sınıflandırma raporu ve karmaşıklık matrisinin oluşturulması
from sklearn.metrics import classification_report,confusion_matrix
```

<p></p>
<b>Sınıflandırma raporu</b> görselleştiricisi, model için kesinlik(precision), geri çağırma(recall), F1(fı-score) ve destek(support) puanlarını görüntüler. Rapor, daha kolay yorumlamayı ve sorun tespitini desteklemek için sayısal puanları renk kodlu bir ısı haritası ile birleştirir. Farklı sınıflandırma raporları arasında sınıflandırma modellerinin kolay karşılaştırmasını kolaylaştırmak için tüm ısı haritaları (0.0, 1.0) aralığındadır.

<ul>
    <li>
        <b>precision</b> bu sınıf arasında kaçının doğru şekilde sınıflandırıldığını gösterir.
    </li>
    <li>
        <b>recall</b> bu sınıftaki tüm eleman sayısı boyunca bu sınıftan kaç tane bulunduğunu gösterir.
    </li>
    <li>
        <b>fı-score</b> hassasiyet ve geri çağırma arasındaki harmonik ortalamadır.
    </li>
    <li>
        <b>support</b> belirli bir sınıfın veri kümesindeki oluşum sayısıdır. 
    </li>
</ul>

```python
print(classification_report(y_test,predictions))
```

<p></p>

<b>Karmaşıklık matrisi</b>, sınıflandırma problemi için bir performans ölçümüdür. Öngörülen ve gerçek değerlerin dört farklı kombinasyonunu içeren bir tablodur.Her ne olursa olsun sınıflandırma tahminleri şu dört değerlendirmeden birine sahip olacaktır:

<ul>
    <li>
        <b>TN(True Negative):</b> Hasta olmayana hasta değil demek.
    </li>
    <li>
        <b>TP(True Positive):</b> Hastaya hasta demek.
    </li>
    <li>
        <b>FN(False Negative):</b> Hasta olana hasta değil demek.
    </li>
    <li>
        <b>FP(False Positive):</b> Hasta olmayana hasta demek.
    </li>
</ul>

<b>Doğruluk(Accuracy)</b>, doğru olarak sınıflandırılan örneklerin yüzdesidir.
<p>accuracy = (TP+TN)/(TP+FP+TN+FN)</p>

<b>Kesinlik(precision)</b>, pozitif olarak tahminlenen değerlerin gerçekten kaç adedinin pozitif olduğunu göstermektedir.
<p>precision = (TP)/(TP+FP)</p>

<b>Duyarlılık(recall)</b>, pozitif olarak tahmin edilmesi gereken işlemlerin ne kadarını pozitif olarak tahmin edildiğini gösteren bir metriktir. “Gerçek pozitiflerin ne kadarı doğru bir şekilde tanımlandı?” sorusunu cevaplar.
<p>precision = (TP)/(TP+FN)</p>



```python
print(confusion_matrix(y_test,predictions))
```
